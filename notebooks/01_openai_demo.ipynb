```python
from dotenv import load_dotenv
import os
from openai import OpenAI
```

```python
load_dotenv()

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)
```

```python
response = client.responses.create(
    model="gpt-4.1-mini",
    input="Hello OpenAI"
)

print(response.output_text)
```
